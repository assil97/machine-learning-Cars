```python
#import the dataset in which we can check the data about how we can 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display 

data=pd.read_csv("http://web.pdx.edu/~gerbing/data/cars.csv")

display(data.head())

#check the length of the data
print("The length of the data",len(data))
```

```python
print(data.describe())
```

```python
pd.scatter_matrix(data,alpha=0.3,figsize=(14,8),diagonal='kde')
```

```python
import seaborn as sns
sns.heatmap(data.corr(),annot=True)
sns.pairplot(data)
```

```python
mpg=data['MPG']
data=data.drop(['MPG'],axis=1)
```

```python
display(mpg.head())
display(data.head())
```

```python
data_dummy=pd.get_dummies(data)
display(data_dummy.head())
```

```python
X=data_dummy
y=mpg
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)
cols=X['Weight']
ax.plot(cols,mpg,'o',color='b')
ax.set_xlabel("Weight")
ax.set_ylabel("MPG")
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)
cols=X['Horsepower']
ax.plot(cols,mpg,'o',color='b')
ax.set_xlabel("Horsepower")
ax.set_ylabel("MPG")
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111)
cols=X['Cylinders']
ax.plot(cols,mpg,'o',color='b')
ax.set_xlabel("Cylinders")
ax.set_ylabel("MPG")
```

```python
from sklearn.cross_validation import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=None)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

print("Linear Regression")
model=LinearRegression()
model_fit=model.fit(X_train,y_train)
model_predict=model_fit.predict(X_train)
print("The r2_score of the model in training is",r2_score(y_train,model_predict))
model_predict=model_fit.predict(X_test)
print("The r2_score os the model in testing is",r2_score(y_test,model_predict))

print("Decision Tree Regressor")
model1=DecisionTreeRegressor()
model1_fit=model1.fit(X_train,y_train)
model_predict_tree=model1_fit.predict(X_train)
print("The r2_score of the model in training is",r2_score(y_train,model_predict_tree))
model_predict_tree1=model1_fit.predict(X_test)
print("The r2_score os the model in testing is",r2_score(y_test,model_predict_tree1))

print("Gradient Boosting Regression")
model2=GradientBoostingRegressor()
model2_fit=model2.fit(X_train,y_train)
model_predict_gradeint=model2_fit.predict(X_train)
print("The r2_score of the model in training is",r2_score(y_train,model_predict_gradeint))
model_predict_gradient1=model2_fit.predict(X_test)
print("The r2_score os the model in testing is",r2_score(y_test,model_predict_gradient1))
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import make_scorer

regressor = GradientBoostingRegressor()
params = {'learning_rate':[0.001,0.01,0.03,1],"n_estimators":[50,100,150,200]}
scoring_fnc = make_scorer(r2_score)
grid = GridSearchCV(estimator=regressor, param_grid=params, scoring=scoring_fnc,)
grid = grid.fit(X_train,y_train)
best_clf=grid.best_estimator_
print(best_clf)
grid_tr=grid.predict(X_train)
grid=grid.predict(X_test)
print("The trainig data is",r2_score(y_train,grid_tr))
print("The testing data is",r2_score(y_test,grid))
```
